# Limpieza de datos

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, accuracy_score
```

```python
data = pd.read_csv('vivino.csv')
```

```python
data
```

## Ya que se han guardado varios vinos con el mismo nombre, bodega y año, se procederá a eliminar los duplicados

```python
# Eliminar filas duplicadas que tienen el mismo nombre y año
data.drop_duplicates(subset=['winery', 'year', 'wine_name'], inplace=True)

# Verificar el resultado
data
```

## Transformación de price a tipo Float y year a tipo Int

```python
# Convertir la columna 'price' a float
data['price'] = data['price'].str.replace('€', '').str.replace(',', '.').astype(float)

# Convertir la columna 'year' a int, reemplazando 'N.V.' por -1
data['year'] = data['year'].replace('N.V.', -1).astype(int)
```

```python
data.value_counts('year')
```

```python
data.describe()
```

## Convertir la columna `wine_type`, `price_quality` y `country` en variables numéricas

```python
# Asignar valores numéricos a los tipos de vino
data['wine_type'] = data['wine_type'].map({'Vino blanco': 1, 'Vino tinto': 2})

# Verificar el resultado
data['wine_type'].value_counts()
```

```python
data['price_quality'].value_counts()
```

```python
# Asignar valores numéricos a los tipos de vino
data.loc[:, 'price_quality'] = data['price_quality'].map({
    'Relación calidad-precio no disponible': 1,
    'Mejor relación calidad-precio en otros vinos': 2,
    'Relación calidad-precio razonable': 3,
    'Buena relación calidad-precio': 4,
    'Excelente relación calidad-precio': 5,
    '¡Increíble relación calidad-precio!': 6})
```

```python
data['price_quality'].value_counts()
```

```python
data['country'].value_counts()
```

```python
# Identificamos los países más comunes
top_countries = data['country'].value_counts().nlargest(3).index

# Creamos una nueva columna 'country_grouped' que agrupa los países menos comunes en 'Other'
data['country_grouped'] = data['country'].apply(lambda x: x if x in top_countries else 'Other')

# comprobar el resultado
print(data['country_grouped'].value_counts())
```

```python
# Asignar valores numéricos a los paises
data.loc[:, 'country_grouped'] = data['country_grouped'].map({
    'Other': 0,
    'Italia': 1,
    'Francia': 2,
    'España': 3})

# Verificar el resultado
print(data['country_grouped'].value_counts())
```

```python
data.info()
```

## Eliminamos columnas innecesarias y con bastantes nulos

```python
data.drop(columns=['flavours', 'Uvas', 'Región', 'Contenido de alcohol'], inplace=True)
```

## Eliminamos los valores nulos y convertimos las columnas `wine_type` y `price_quality` y `country_grouped` a tipo int

```python
data.dropna(inplace=True)

## Convertimos las columnas wine_type y price_quality a tipo int

data['wine_type'] = data['wine_type'].astype(int)
data['price_quality'] = data['price_quality'].astype(int)
data['country_grouped'] = data['country_grouped'].astype(int)
```

```python
data.info()
```

```python
# Guardar el DataFrame limpio en un archivo CSV
data.to_csv('vivino_clean.csv', index=False)
```
