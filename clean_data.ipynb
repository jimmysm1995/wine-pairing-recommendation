### Manipulación de datos nulos

```python
# Vemos si hay datos nulos
print(data.isna().sum())
```

#### Rellenar registros con la media de esa columna
Voy a rellenar los registros nulos de la columna `Contenido de alcohol` con la media de esta columna ya que la mayoria de los vinos tienen un rango de 12%-15% y la media es de 13,91%

```python
# Eliminar el símbolo '%' y convertir la columna 'Contenido de alcohol' a tipo numérico
data['Contenido de alcohol'] = data['Contenido de alcohol'].astype(str).str.replace('%', '', regex=False).astype(float)

# Calcular la media y redondearla a dos decimales
media_alcohol = data['Contenido de alcohol'].mean()
media_alcohol = round(media_alcohol, 2)

# Imprimir la media redondeada
print(f"La media del contenido de alcohol es: {media_alcohol}")
```

```python
# Rellenar los valores NaN en 'Contenido de alcohol' con la media calculada
data['Contenido de alcohol'] = data['Contenido de alcohol'].fillna(media_alcohol)

# Verificar si se han reemplazado los valores NaN, debería salir 0
print(data['Contenido de alcohol'].isna().sum())
```

#### Eliminar los demás registros que tienen otros campos nulos
Los siguientes registros que debemos eliminar ya que no podemos rellenar los datos corresponden con las columnas `pairing`, `flavours`, `price_quality` y `Uvas`

Después comprobamos si quedan registros con nulos

```python
data = data.dropna(subset=['pairing', 'flavours', 'price_quality', 'Uvas'])

# Vemos si hay datos nulos
print(data.isna().sum())
```

## Una vez eliminados los datos duplicados y nulos, procedemos a convertir los tipos de columnas en los correctos
- Convertir los datos de la columna `price`a tipo float
- Convertir los datos de la columna `year`a tipo int
- Codificar los datos de la columna `wine_type` a una etiqueta numérica (`Label Encoding`)
- Asignarle un valor numérico a los valores de la columna `price_quality`, es decir, mapear los datos
- Agrupar los paises, voy a seprar los 3 que más comunes son y los demás en otro grupo, es decir, quedarán 4 grupos y aplicar `one-hot encoding`

```python
# Convertir la columna 'price' a float
data['price'] = data['price'].str.replace('€', '').str.replace(',', '.').astype(float)

# Eliminar filas con NaN en la columna 'year'
data = data.dropna(subset=['year'])

# Convertir 'year' a int
data['year'] = data['year'].astype(int)
```

```python
# Asignar valores numéricos a los tipos de vino
data['wine_type'] = data['wine_type'].map({'Vino blanco': 1, 'Vino tinto': 2})
```

```python
# Asignar valores numéricos a los tipos de vino
# 1 representa la peor relación calidad-precio (valor más bajo)
# 6 representa la mejor relación calidad-precio (valor más alto)
data['price_quality'] = data['price_quality'].str.strip().map({
    'Relación calidad-precio no disponible': 1,
    'Mejor relación calidad-precio en otros vinos': 2,
    'Relación calidad-precio razonable': 3,
    'Buena relación calidad-precio': 4,
    'Excelente relación calidad-precio': 5,
    '¡Increíble relación calidad-precio!': 6})
```

```python
data['price_quality'].value_counts()
```

```python
# Vemos los paises que hay en el dataset
data['country'].value_counts()
```

```python
# Identificamos los países más comunes
top_countries = data['country'].value_counts().nlargest(3).index

# Creamos una nueva columna 'country_grouped' que agrupa los países menos comunes en 'Other'
data['country_grouped'] = data['country'].apply(lambda x: x if x in top_countries else 'Other')

# comprobar el resultado
print(data['country_grouped'].value_counts())
```

```python
# Aplicar one-hot encoding a la columna 'country_grouped'
country_encoded = pd.get_dummies(data['country_grouped'], prefix='country')

# Unir las nuevas columnas al DataFrame original
data = pd.concat([data, country_encoded], axis=1)

# Eliminamos la columna inecesaria
data = data.drop(columns=['country_grouped'])

# Verificar el resultado
data.head()
```
